```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

```python
train_df = pd.read_csv("/content/Salary.csv")
```

```python
train_df
```

```python
print(train_df.head())
print(train_df.isnull().sum())
print(train_df.describe())
print(train_df.info())
```

```python
x = train_df['YearsExperience']
y = train_df['Salary']
```

```python
w = 0.0
b = 0.0

lr = 0.01
epochs = 1000
```

```python
m = len(x)
for epoch in range(epochs):
  y_pred = w*x+b
  error = y_pred - y

  dw = (2/m)*np.dot(error,x)
  db = (2/m)*np.sum(error)

  w -= lr*dw
  b -= lr*db

  if epoch%100 == 0:
    loss = np.mean(error**2)
    print(f"Epoch {epoch} : Loss = {loss :.2f}, w = {w:.2f}, b = {b:.2f}")

    # Plot original data
    plt.scatter(x, y, color='blue', label='Data points')

    # Plot predicted line
    y_line = w * x + b
    plt.plot(x, y_line, color='red', label='Regression Line')

    plt.xlabel('Hours Studied')
    plt.ylabel('Exam Score')
    plt.title('Linear Regression Example')
    plt.legend()
    plt.grid(True)
    plt.show()
```

```python
# Plot original data
plt.scatter(x, y, color='blue', label='Data points')

# Plot predicted line
y_line = w * x + b
plt.plot(x, y_line, color='red', label='Regression Line')

plt.xlabel('Hours Studied')
plt.ylabel('Exam Score')
plt.title('Linear Regression Example')
plt.legend()
plt.grid(True)
plt.show()
```

```python
new_study = np.array([5])
prediction = (np.dot(new_study,w)+b)
print(f"Prediction : {prediction}")
```
